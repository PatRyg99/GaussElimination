## Patryk Rygiel

```julia
using DelimitedFiles
using Plots
```

```julia
gauss_stats = float(open(readdlm, "benchmark/GaussStats.txt"))
gauss_choice_stats = float(open(readdlm, "benchmark/GaussChoiceStats.txt"))
gauss_lu_stats = float(open(readdlm, "benchmark/GaussLuStats.txt"))
gauss_lu_choice_stats = float(open(readdlm, "benchmark/GaussLuChoiceStats.txt"))

gauss_stats = gauss_stats[sortperm(gauss_stats[:, 1]), :]
gauss_choice_stats = gauss_choice_stats[sortperm(gauss_choice_stats[:, 1]), :]
gauss_lu_stats = gauss_lu_stats[sortperm(gauss_lu_stats[:, 1]), :]
gauss_lu_choice_stats = gauss_lu_choice_stats[sortperm(gauss_lu_choice_stats[:, 1]), :]
```

## Plots

```julia
data = [
    gauss_stats[:, 2], gauss_choice_stats[:, 2], gauss_lu_stats[:, 2], gauss_lu_choice_stats[:, 2]
]

plot(
    gauss_stats[:, 1], 
    data, 
    title="Porównanie czasu dla L=4",
    label=["Gauss" "Gauss + częsciowy wybór" "Gauss + LU" "Gauss + LU + częściowy wybór"],
    lw=3,
    legend=:topleft
)

ylabel!("Czas [s]")
xlabel!("Wymiar macierzy A")

#savefig("time_plot.png")
```

```julia
data = [
    gauss_stats[:, 3], gauss_choice_stats[:, 3], gauss_lu_stats[:, 3], gauss_lu_choice_stats[:, 3]
]

plot(
    gauss_stats[:, 1], 
    data, 
    title="Porównanie błędu względnego dla L=4",
    label=["Gauss" "Gauss + częsciowy wybór" "Gauss + LU" "Gauss + LU + częściowy wybór"],
    lw=3,
    legend=:topleft
)

ylabel!("Błąd względny")
xlabel!("Wymiar macierzy A")

#savefig("error_plot.png")
```
